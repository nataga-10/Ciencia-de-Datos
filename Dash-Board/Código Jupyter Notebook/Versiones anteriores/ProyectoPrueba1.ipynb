```python
import requests

def get_vulnerability_score(score):
    if score is None:
        return "Desconocida"
    elif score == 0:
        return "Nula"
    elif 0.1 <= score <= 3.9:
        return "Baja"
    elif 4.0 <= score <= 6.9:
        return "Media"
    elif 7.0 <= score <= 8.9:
        return "Alta"
    elif 9.0 <= score <= 10.0:
        return "Crítica"
    else:
        return "Desconocida"

cve_id = "CVE-2023-2464"
url = f'https://cve.circl.lu/api/cve/{cve_id}'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    cvss_score = data.get('cvss', 0)
    vulnerability_level = get_vulnerability_score(cvss_score)
    
    print(f"ID CVE: {cve_id}")
    print(f"Fecha de Modificación: {data.get('last-modified')}")
    print(f"Fecha de Publicación: {data.get('Published')}")
    print(f"CVSS: {cvss_score}")
    print(f"Vector CVSS: {data.get('cvss-vector')}")
    print(f"Impacto: {data.get('impact')}")
    print(f"Vulnerabilidad: {vulnerability_level}")

    affected_products = data.get('vulnerable_product', [])
    if isinstance(affected_products, list):
        print("\nProductos afectados:")
        for product_info in affected_products:
            if isinstance(product_info, dict):
                product = product_info.get('product', '')
                vendor = product_info.get('vendor', '')
                version = product_info.get('version', '')
                print('  - Product:', product)
                print('    Vendor:', vendor)
                print('    Version:', version)
    else:
        print('No se encontraron datos de afectación para la CVE especificada.')
else:
    print("Error al obtener los datos. Código de estado:", response.status_code)
```
