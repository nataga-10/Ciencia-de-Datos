```python
import plotly.graph_objects as go
from pymongo import MongoClient

uri = "mongodb+srv://LocoMe:mERDUx9dHYzEMZuk@bryannathi.y0j00jj.mongodb.net/?retryWrites=true&w=majority&appName=BryanNathi"

# Conexión a MongoDB
cliente = MongoClient(uri)
db = cliente["BryanNathi"]
coleccion = db["CVE"]

# Obtener los datos
documentos = coleccion.find()

# Contadores
computadores = {}
cves = {}
programas = {}
fechas = {}

# Procesar los datos
for documento in documentos:
    nombre_equipo = documento.get("host-name", "")
    programas_doc = documento.get("programas", [])
    for programa in programas_doc:
        cve = programa.get("ID_CVE", "")
        detalles_cve = programa.get("detalles_CVE")
        if detalles_cve:  # Verificar si detalles_cve no es None
            score = detalles_cve.get("cvss", 0)
            fecha_publicacion = detalles_cve.get("Published", "")
            
            # Conteo por computador
            computadores[nombre_equipo] = computadores.get(nombre_equipo, 0) + 1
            
            # Conteo de CVE
            cves[cve] = cves.get(cve, 0) + 1
            
            # Conteo por programa
            nombre_programa = programa.get("nombre_programa", "")
            programas[nombre_programa] = programas.get(nombre_programa, [])
            if score is not None:  # Filtrar valores None
                programas[nombre_programa].append(score)
            
            # Conteo por fecha
            fechas[fecha_publicacion] = fechas.get(fecha_publicacion, 0) + 1

# Imprimir para debugear
print("Computadores:")
print(computadores)
print("\nCVEs:")
print(cves)
print("\nProgramas:")
print(programas)
print("\nFechas:")
print(fechas)

# Calcular el score promedio por programa y clasificar programas por nivel de vulnerabilidad
score_ranges = [
    (0, "Nula"),
    (3.9, "Baja"),
    (6.9, "Media"),
    (8.9, "Alta"),
    (10, "Crítica")
]

programas_por_nivel = {nivel: [] for _, nivel in score_ranges}

for programa, scores in programas.items():
    scores_filtrados = [score for score in scores if score is not None]
    if scores_filtrados:  # Verificar si la lista no está vacía
        promedio_score = sum(scores_filtrados) / len(scores_filtrados)
        for threshold, nivel in score_ranges:
            if promedio_score <= threshold:
                programas_por_nivel[nivel].append(programa)
                break

# Imprimir programas por nivel de CVSS Score
print("\nProgramas por nivel de CVSS Score:")
print(programas_por_nivel)

# Crear las gráficas
fig_computadores = go.Figure(data=[go.Bar(x=list(computadores.keys()), y=list(computadores.values()))])
fig_computadores.update_layout(title="Computadores más vulnerables")

fig_top_cves = go.Figure(data=[go.Bar(x=list(cves.keys()), y=list(cves.values()))])
fig_top_cves.update_layout(title="Top CVEs más encontrados")

fig_programas_vulnerables = go.Figure(data=[go.Bar(x=list(programas_por_nivel.keys()), y=[len(programas_por_nivel[nivel]) for nivel in programas_por_nivel.keys()])])
fig_programas_vulnerables.update_layout(title="Programas más vulnerables por nivel de CVSS Score")

fig_top_fechas = go.Figure(data=[go.Bar(x=list(fechas.keys()), y=list(fechas.values()))])
fig_top_fechas.update_layout(title="Top CVEs por fecha de publicación")

fig_contador_cves = go.Figure(data=[go.Bar(x=["CVEs encontradas"], y=[len(cves)])])
fig_contador_cves.update_layout(title="Contador de todas las CVE encontradas")

# Mostrar las gráficas
fig_computadores.show()
fig_top_cves.show()
fig_programas_vulnerables.show()
fig_top_fechas.show()
fig_contador_cves.show()
```

```python
import plotly.graph_objects as go
from pymongo import MongoClient

uri = "mongodb+srv://LocoMe:mERDUx9dHYzEMZuk@bryannathi.y0j00jj.mongodb.net/?retryWrites=true&w=majority&appName=BryanNathi"

# Conexión a MongoDB
cliente = MongoClient(uri)
db = cliente["BryanNathi"]
coleccion = db["CVE"]

# Contador de CVEs por computador
computadores = {}

# Obtener los datos y contar CVEs por computador
documentos = coleccion.find()
for documento in documentos:
    nombre_equipo = documento.get("host-name", "")
    programas_doc = documento.get("programas", [])
    for programa in programas_doc:
        cve = programa.get("ID_CVE", "")
        if nombre_equipo:
            computadores[nombre_equipo] = computadores.get(nombre_equipo, 0) + 1

# Verificar si hay datos para graficar
if computadores:
    # Obtener el computador más vulnerable (mayor cantidad de CVEs)
    computador_mas_vulnerable = max(computadores, key=computadores.get)
    
    # Crear el gráfico de torta
    fig_computador_vulnerable = go.Figure(data=[go.Pie(labels=list(computadores.keys()), values=list(computadores.values()))])
    fig_computador_vulnerable.update_layout(title=f"Computador más vulnerable: {computador_mas_vulnerable}")

    # Mostrar el gráfico
    fig_computador_vulnerable.show()
else:
    print("No se encontraron datos de CVEs por computador para graficar.")
```
